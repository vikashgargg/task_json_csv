```python
import pandas as pd
import json
```

```python
with open('sample_record_json_format.json') as f:
    d=f.read()
    d=json.loads(d)
    df=pd.json_normalize(d)
df.to_csv('sample.csv',index=False)    
```

```python
df1=df.explode('payload')
```

```python
df2=pd.json_normalize(json.loads(df1.to_json(orient="records")))
```
